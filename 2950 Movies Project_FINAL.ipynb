```python
def stat_summary(df, group):

    """
        Function that prints out the average, median, and standard deviation of the audience score group by 'group'  
    """
    df_group = df.groupby(by = group)
    print('Average audience ratings by', group, ":")
    print(df_group[["audience_score"]].mean())
    print("--------")
    print('Median audience ratings by ', group, ":")
    print(df_group[["audience_score"]].median())
    print("--------")
    print('Standard deviation of audience ratings by', group, ":")
    print(df_group[["audience_score"]].std())
```

Although the average audience ratings are important numbers to look at, we are focusing our data analysis on genres that have had a positive trend in audience ratings over the recent years. That being said, a genre that possibly has a high average audience ratings could project a negative trend over the years. Hence, we would want to calculate the slopes of the top 5 genres to see which genres have a positive trend in their audience ratings, which would help us in determining which genre to focus on.


```python
#function to print out coefficients of genre audience ratings over the years 

def slope_coefficient(df, genre):
    
    genre= df[df["genre"]== genre]
    x_year = genre["release_year"].values.reshape(-1,1)
    y_rating = genre["audience_score"]
        
    genre_model = LinearRegression().fit(x_year,y_rating)
    
    return genre_model.coef_
```

```python
#running function to print slope of each genre 

list_genre = after_2000_most.genre.unique()
list_genre

for x in list_genre:
     print(x)
     slope = slope_coefficient(after_2000_most, x) 
     print(slope)
```

## Individual Analysis on Quantitative Variables 
After choosing comedies to be our genre that we conduct our analysis on, we can now start looking at the individual significance of the quantitative variables: release year, duration, revenue, and budget. In other words, we would like to see if each variable has a significant impact in audience scores. For this linear regression analysis, we looked into ordinary least squares regression (OLS), which minimizes the sum of squared differences between the independent and dependent variables. We also created a function that helps us visualize the linear regression model. 

```python
# create dataset of comedy movies
comedies_df = genre_df2[genre_df2["genre"] == "Comedies"]
comedies_df = comedies_df[comedies_df["release_year"]>=2000]
```

```python
def lin_reg(df, var):
    
    """""
    Function that creates a linear regression between 'var' and audience score based. It then creates a 
    visual regression with both true and predicted values based on the sklearn linear regression. 
    It then prints out OLS for further analysis on the significance of the linear regression. 
    Finally, it creates a residual plot of between the predicted and true values. 
    """""
    
    x_var = df[var].values.reshape(-1,1)
    audience_score = df["audience_score"].values

    print("Indicator variable: ", var)
    
    #Creating a Linear Regression based on the inputed dataframe and x variable 
    reg = LinearRegression().fit(x_var, audience_score)
    prediction = reg.predict(x_var)
    
    #Creates a visual regression model with both true and predicted values
    plt.scatter(x_var, audience_score, label = "True values") 
    plt.plot(x_var,prediction, color="orange")
    plt.scatter(x_var, prediction, label = "Predictions based on Linear Regression model")
    plt.legend()
    plt.title("Audience Rating Score vs. " + var + " colored by predicted or true values")
    plt.xlabel(var)
    plt.ylabel("Audience Rating Score")
    plt.show()

    # OLS on variable and audience rating
    x_var = sm.add_constant(x_var)
    OLS_model = sm.OLS(audience_score, x_var).fit()
    print(OLS_model.summary())

    # residual plot of variable and audience rating
    sns.residplot(data = df, x= var, y= 'audience_score')
    plt.show()
```

### Linear Regression for Release Year vs Audience Score

```python
#Linear Regression for release year verus audience score

lin_reg(comedies_df, "release_year")
```

From the results demonstrated above, we can conclude that there is a general positive relationship between the release year and audience rating. The model predicts that as each year increases by one year, the standardized audience rating decreases by 0.8784; The p-value of release year is 0.002 < 0.05, thus this beta value is statistically significant in indicating that the release year is related to the changes in audience rating. It is worth noting that the r-squared value of 0.015 is extremely low, which means that the model does not explain the prediction well. This is likely due to the lack of other variables since this is only the individual regression analysis. 

As shown above, the residuals for audience ratings on release year are spread out randomly with no specific trend. This suggests that the linear model is appropriate without favoring certain points.

# Multivariable Analysis

From the analysis of individual variables, it is very difficult to only attribute a movie's success to only one variable. Therefore, we delve into multivariate regression analysis to see how a combination of quantitative variables would impact audience score in a statistically significant way. 

We first designed a function that takes in  train/test sets to create a linear regression based on the imputed arguments. The function also calculates root mean squared error (RMSE) and mean absolute error (MAE) to serve as a metric to evaluate the fit of the regression to the model. 

For every multivariable regression, we will use OLS to calculate the test statistic to help evaluate the significance of each imputed variable. 



```python
# regression function
def run_regression(li, train, test, predicted):
    
    """"
    Function that takes in train and test set to create a linear regression based on the imputed arguments. 
    The function also calculates root mean squared error (RMSE) and mean absolute error (MAE) to 
    serve as a metric to evaluate the fit of the regression to the model. 

    """
    
    
    print("Variables: " + str(li))
    X = pd.DataFrame()
    X_test = pd.DataFrame()
    for v in li:
        X[v] = train[v]
        X_test[v] = test[v]
    model = LinearRegression().fit(X, train[predicted])
    print("Coeffcient of Train Data: " + str(model.coef_))
    print("Intercept: " + str(model.intercept_))
    
    # train RMSE
    y_train = model.predict(X)
    mse_train = ((train[predicted] - y_train)**2).mean()
    rmse_train = np.sqrt(mse_train)
    print("Train RMSE: " + str(rmse_train))
    
    # test RMSE
    y_test = model.predict(X_test)
    mse_test = ((test[predicted] - y_test)**2).mean()
    rmse_test = np.sqrt(mse_test)
    print("Test RMSE: " + str(rmse_test))
    
    # train MAE
    mae_train = np.absolute(train[predicted] - y_train).mean()
    print("Train MAE: " + str(mae_train))
    
    # test MAE
    mae_test = np.absolute(test[predicted] - y_test).mean()
    print("Test MAE: " + str(mae_test))
    
```

## Multivariable analysis on all quantitative variables


```python
data_train, data_test = train_test_split(country_num, test_size=0.1, random_state=5)

list_input = ['release_year', 'MA_rating_rank', 'budget', 'country_rank', 'duration', 'revenue']
run_regression(list_input, data_train, data_test, 'audience_score')

multi_list = [ 'MA_rating_rank', 'budget', 'country', 'duration']

multi_x = country_num[list_input]

multi_x = sm.add_constant(multi_x)
multi_summary = sm.OLS(country_num["audience_score"], multi_x).fit()
print(multi_summary.summary())
```

We trained a multivariate linear regression that takes in the release year, Motion Picture Association ratings, budget, production countries, and duration. We chose to leave revenue out of the multivariable regression because it is not a factor or decision that is put in while making the movie. 

From the results exhibited above, the model suggests that there is a general negative relationship between release year, budget, MPA ratings, and production country. There is a general positive relationship between duration and audience score. Of the 5 variables in the multivariate regression, only duration and budget had p values that are smaller than the alpha value of 0.05. Hence, this suggests that budget and duration are the only two variables that have a statistically significant correlation on the audience score. It is worth noting that the r-squared value of 0.154 is low, which means that the model does not explain the prediction well. This is likely due to the large variance from all the data and having so many variables. 

Using evaluation metrics of train and test sets, we can see that the test RMSE and MAE are similar but lower than the train set, meaning that this is a relatively good fitting model.

## Bootstrap analysis to find confidence intervals 


```python
#bootstrap analysis 

budget_coeff = []
duration_coeff = []
for i in range(1000):
    sample = country_num.sample(n= comedies_df.shape[0], replace = True)
    x_boot  =  sample[['budget', 'duration']].values
    score_boot = sample["audience_score"]
    reg_boot = LinearRegression().fit(x_boot,score_boot)
    budget_coeff.append(reg_boot.coef_[0])
    duration_coeff.append(reg_boot.coef_[1])
    
    
print("Budget:")
print('Expected value: ',np.mean(budget_coeff))
print('Standard Error: ',np.std(budget_coeff))
print('95% Confidence Interval for regression coefficient: {',np.percentile(budget_coeff,0.5),'to',np.percentile(budget_coeff,97.5),'}')
print("---------")
print("Duration:")
print('Expected value: ',np.mean(duration_coeff))
print('Standard Error: ',np.std(duration_coeff))
print('95% Confidence Interval for regression coefficient: {',np.percentile(duration_coeff,0.5),'to',np.percentile(duration_coeff,97.5),'}')
    


fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2)
ax0.hist(budget_coeff)
ax0.set_title('Budget Coefficient')
ax0.set_xlabel('Coefficients')
ax0.set_ylabel('Frequencies')
ax0.axvline(np.mean(budget_coeff), color= "r")



ax1.hist(duration_coeff)
ax1.set_title('Duration Coefficient')
ax1.set_xlabel('Coefficients')
ax1.set_ylabel('Frequencies')
ax1.axvline(np.mean(duration_coeff), color= "r")


      
fig.tight_layout()

```

To further evaluate the results, we wanted to find the confidence intervals of our results. Thus, we used a bootstrapping method to stimulate resampling with replacement to create multiple samples from the given dataset. 


For **budget**, from the results above, through the 1000 bootstrap samples, we find that the model predicts with every increase in $100 million in revenue for comedy movies, the average audience score decreases approximately 8 points.  The model predicts the 95% confidence interval of the true budget coefficient is approximately between -2 to -16 points. Given that the confidence interval does not contain zero, this indicates that the coefficient for budget is significant and indeed negatively correlated with audience score. 



For **duration**, from the results above, through the 1000 bootstrap samples, we find that the model predicts with every increase in one minute for comedy movies, the average audience score increases approximately 0.4 to 0.5 points.  The model predicts the 95% confidence interval of the true duration coefficient is approximately between 0.1 to 0.6 points. Given that the confidence interval does not contain zero, this indicates that the coefficient for duration is significant and indeed positively correlated with audience score. 


# Analysis on Revenue versus Rotten Tomato Audience Score

After observing the importance of budget and duration to a comedy movie in terms of audience score, we wanted to further explore the impact of audience score on the financial performance of a movie. In particular, since the financial success of a movie is often associated with high ratings and deemed as a “hit movie”, we wanted to see whether it is worth prioritizing audience score when producing a movie. In other words, is the reception of the movie correlated with the movie? 

Thus, we decide to use bootstrapping to estimate the impact of audience score on revenue by creating 1000 samples that draw with replacement. We first start the analysis using comedy movies revenue in the past 20 years to see the impact of revenue on ratings. We then continue the analysis on all movies regardless of their genre to see whether the revenue analysis still holds true when applied to a larger set of movies. 


## a. Revenue and Audience Score Analysis for Comedy Movies

```python
#print the average revenue of comedy movies 
print("The average audience score of comedy movies, according to the dataset, is:" ,np.mean(comedies_df["audience_score"]))
print("The average revenue of comedy movies, according to the dataset, is:" ,np.mean(comedies_df["revenue"]))
```

```python
#bootstrap analysis on revenue of comedy movies 


coeff = []
intercept = []
mses = []
r_squared = []

for i in range(1000):
    sample = comedies_df.sample(n= comedies_df.shape[0], replace = True, random_state=i*100)
    x_boot  =  sample["audience_score"].values.reshape(-1,1)
    score_boot = sample["revenue"]
    reg_boot = LinearRegression().fit(x_boot,score_boot)
    predict_score = reg_boot.predict(x_boot)
    
    coeff.append(reg_boot.coef_[0])
    intercept.append(reg_boot.intercept_)
    r_squared.append(reg_boot.score(x_boot,score_boot))
    
    mses.append(np.square(np.subtract(score_boot,predict_score)).mean())


print('Expected value: ',np.mean(coeff))
print('Standard Error: ',np.std(coeff))
print('95% Confidence Interval for regression coefficient: {',np.percentile(coeff,2.5),'to',np.percentile(coeff,97.5),'}')
    
plt.hist(coeff)    
plt.title('Revenue Coefficient')
plt.xlabel('Coefficients')
plt.ylabel('Frequencies')
plt.axvline(np.mean(coeff), color= "r")
```

From the results above, through the 1000 bootstrap samples, we find that the model predicts with every increase in audience score, the mean revenue will increase by approximately 1,000,000 (will vary due to random sampling). The intercept is around 80 million, which brings up a limitation of the dataset as there must not have been movies with less than 80 million in revenue. The model predicts the 95% confidence interval is approximately between -0.5 to 3 million dollars . However, given that the confidence interval contains zero, this indicates that the coefficient above is not significant. 


```python
xp = np.linspace(0,100,100)

sns.regplot(data= comedies_df,x='audience_score',y='revenue',ci=95,scatter=False,color='r')
plt.scatter(x=comedies_df["audience_score"],y=comedies_df["revenue"],color='b')
plt.plot(xp,np.mean(coeff)*xp+np.mean(intercept)+2*np.mean(np.sqrt(mses)),'-.g',linewidth=2)
plt.plot(xp,np.mean(coeff)*xp+np.mean(intercept)-2*np.mean(np.sqrt(mses)),'-.m',linewidth=2)
plt.title('Linear Regression Fit of Comedy Movie Audience Score versus Revenue', fontsize=18)
plt.xlabel('Audience Score', fontsize=14)
plt.ylabel('Comedy Movie Revenue', fontsize=14)
plt.legend(['Model','95% CI','Comedy Movies','Model+2RMSE','Model-2RMSE'],loc='lower right')
plt.show()
print("Average R squared value of the Model:", np.mean(r_squared))
```

This model was created by plotting the true data points from the comedy movies with the regression line created from bootstrapping coefficient analysis. Because RMSE can approximate the gaussian noise, we multiplied the RMSE (average obtained from the bootstrapping analysis)  by 2 to approximate upper and lower bounds that demonstrates 95% spread of the data. 

From the model, it can be seen that the highest grossing movie on the linear regression has a fairly high audience score; however, due to the limited data from our dataset, this data point acts more like an outlier. The data points around the average comedy audience score of 60 demonstrates the wider spread of revenue, ranging from 1 million to 200 million. Because of this large spread and also the small r-squared value, this helps visualize why audience score might not be a good predictor of comedy movie revenue. 

The data also shows heteroskedasticity, so the revenue data was put into a logarithmic transformation.

```python
xp = np.linspace(0,100,100)

sns.regplot(data= movies_df,x='audience_score',y='log_revenue',ci=95,scatter=False,color='r')
plt.scatter(x=movies_df["audience_score"],y=movies_df["log_revenue"],color='b')
plt.plot(xp,np.mean(coeff_2)*xp+np.mean(intercept_2)+2*np.mean(np.sqrt(mses_2)),'-.g',linewidth=2)
plt.plot(xp,np.mean(coeff_2)*xp+np.mean(intercept_2)-2*np.mean(np.sqrt(mses_2)),'-.m',linewidth=2)
plt.title('Linear Regression Fit of Audience Score versus Revenue of All Movies', fontsize=18)
plt.xlabel('Audience Score', fontsize=14)
plt.ylabel('Revenue', fontsize=14)
plt.legend(['Model','95% CI','Movies','Model+2RMSE','Model-2RMSE'],loc='lower right')
plt.show()
print("Average R squared value of the Model:", np.mean(r_squared_2))
```

The resulting regression of log(revenue) vs. audience score for all movies showed reduced and almost no heteroskedasticity. The model predicts the 95% confidence interval is approximately between -0.00669 and 0.0183. However, given that the confidence interval contains zero, this indicates that the coefficient above is  ***not significant***. Furthermore, it is worth pointing out that the r-squared value is even smaller in the linear regression above, indicating that this model might not be a good fit for audience score to predict revenue.

# Limitations
## A. Limitations of the Dataset
The first limitation the data has is that there is only data up to early 2019, which means this data is not inclusive of the latest movies released. Hence, we would be making a generalization of the population based on data collected two years ago, knowing that trends are changing rapidly because of social media effects and also changes in the audience population. This is also a large limitation because we are not able to analyze the potential changes after COVID. In other words, people's behaviors might have changed that lead to differences in movie production that change the definition of movie popularity.


Another limitation is that Rotten Tomatoes might not be a representative measure of customer satisfaction for movies. It would also be beneficial to look at streaming hours on Netflix or IMDB ratings if the data was accessible to us. However, Netflix only began releasing its streaming hours starting from 2021 and not in the form of a dataset. Because we also only chose to look at Rotten Tomato audience ratings, this might only be representative of a subset of audience with niche taste as movie experiences are very subjective. Furthermore, we have no control over whether the audience is giving their honest opinions or if they are swayed by external factors.  Therefore, to limit rater bias, more evaluation metrics of how “successful” a movie is should be taken into account for future analysis. 


Furthermore, as mentioned in the Appendix, the number of reviews may vary per movie. Let’s say a movie had a 10.0 rating but only 1 audience review versus a movie with a million reviews and a 7.0 review. It would be hard to gauge general audience satisfaction based on one audience’s opinion despite having a high rating. Although we attempted to resolve this by standardizing the number of reviews, we ran into the issue of how we define movies with “little” movie reviews. We ended up referencing the IMDB cutoff, which specifies that their minimum vote needed to be listed in Top 250 is 3000 votes [https://www.imdb.com/chart/top/]. However, a large limitation to be noted here is that this number comes from IMDB, which may have a different metric compared to Rotten Tomatoes.


This dataset might also not be the most comprehensive dataset of all movies that were released as many titles had to be removed due to the lack of information on revenue and budget. Therefore, our dataset only had around 130 comedy movies for analysis, which might not be representative of all the comedy movies. 


## b. Limitations in the Analysis 
One of the largest limitations in our analysis is that linear regressions are very sensitive to outliers. Since many movies may fall out of the general trend due to various factors, the linear regression is easily swayed by those data points. For example, there might be movies which had a really high budget and had low audience reception doesn’t mean that all movies that have a high movie budget will always do poorly. There might be other factors at play that we were not able to analyze. 


This brings us to another limitation where we initially had a lot of categorical variables. In particular, the data from the director, maturity ratings, and the countries that the movies were produced in were all categories and they each had multiple categories (ex. Within the genres, there are comedies, thrillers, dramas, etc.). To make it more complex, each movie might also fit into multiple categories; for example, a movie can be both comedy and horror. This made our analysis less clear as there was not a clearer cut between each category. This also eliminated k-clustering analysis and also made multivariable analysis very tricky.  We worked around this limitation by first narrowing down one genre and conducting multivariable analysis on quantifiable variables within a genre, however, we had to sacrifice variables such as director or leading cast, which might also provide valuable information regarding the success of the movie. Another variable that we sacrifice is the genre analysis and whether the traits vary between different genres. Our analysis only focuses on comedy movies but we were not able to see any of the trends in other genres such as horror or thrillers. In the future, one can look into a dataset that provides more information about the director or genre to see if any traits can be identified. 


Even though we tried to resolve this by turning the values into numerical values, we ran into another issue where our sample sizes for each category might be too small. For example, there were really only four maturity audience ratings that had enough information on each title to be included in the analysis. Although the train RMSE/MSE and test RMSE/MSE are very similar, many categories lack data points for a more concrete answer. Hence, for this conclusion to be generalizable and relevant, more variables should be analyzed in order for a more comprehensive analysis of what is truly important behind creating a movie.
